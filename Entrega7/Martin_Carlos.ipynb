```python
from pynini import *
import os
```

```python
# Abrimos el archivo

archivo = open("texto.html")
```

```python
def modificaLineas(linea):
    
    chars = [chr(i) for i in range(1,91)] + [r"\[", r"\\", r"\]"] + [chr(i) for i in range(94,256)] 
    sigma_star = union(*chars).closure() 

    singular_map = union( string_map([("<h2>", "## ")]),string_map([("</h2>", "")]),string_map([("<html>","")]),string_map([("<h3>", "### ")]),string_map([("</h3>", "")]),string_map([("<h1>", "# ")]),string_map([("</h1>", " ")]),string_map([("<h4>", "#### ")]),string_map([("</h4>", "")]),string_map([("<h5>", "##### ")]),string_map([("</h5>", "")]),string_map([("<h6>", "###### ")]),string_map([("</h6>", "")]),string_map([("<br>", "  \n")]),string_map([("</br>", "")]),string_map([("<i>", "*")]),string_map([("</i>", "*")]),string_map([("<b>", "**")]),string_map([("</b>", "**")]),string_map([("<ul>", "")]),string_map([("<li>", "- ")]),string_map([("</li>", "\n")]),string_map([("<hr>", "---")]),string_map([("</html>", "")]),string_map([("<p>", "")]),string_map([("</p>","\n\n")]),string_map([("</ul>", "\n")]),string_map([("\n", "\t")]),string_map([("<a href=\"","\[(")]), string_map([("\">",")")]), string_map([("</a>","\]")]))
    
                
    rewrite = cdrewrite(singular_map, "", "", sigma_star)

    
    # Printeo la linea que recibe como parametro del archivo HTML
    
    print("Cadena original HTML: ",linea)
    
    # Probamos a modificar la linea
    
    cadena_etiquetada = compose(linea, rewrite).string()
    
    # Si contiene un corchete [ ] o ( ), va a tener enlace por tanto realizamos los cambios que se producen en el if para rotarlos, sino como una cadena normal
    
    if('[' in cadena_etiquetada and ']' in cadena_etiquetada and '(' in cadena_etiquetada and ')' in cadena_etiquetada):
        
        # Limite1 y limite2 representa donde empieza y acaba, respectivamente el enlace
        
        limite1 = cadena_etiquetada.index('(')
        limite2 = cadena_etiquetada.index(')')+1

        # Donde termina la cadena que no es enlace y hay que intercambiar con el enlace
        
        limite3 = cadena_etiquetada.index(']')+1
        
        # Definimos las cadenas
        
        enlace = cadena_etiquetada[limite1:limite2]
        cadenaCambio = cadena_etiquetada[limite2:limite3]
        
        # Ahora para dar la vuelta al enlace y al Universidad de Valladolid, utilizo la funcion replace, que sustituye una cosa por otra
        # primero sustituyo Universidad de valladolid por donde deberia de ir el enlace, luego coloco bien Universidad de Valladolid
        # y por ultimo pongo el enlace donde debe de ir en "aquivaelenlace"
    

        cadena_etiquetada2 = cadena_etiquetada.replace( cadenaCambio,"aquivaelenlace")
    
        cadena_etiquetada2 = cadena_etiquetada2.replace( enlace ,cadenaCambio)
    
        cadena_etiquetada2 = cadena_etiquetada2.replace("aquivaelenlace", enlace)
        
        print("Cadena formato MARKDOWN: ", cadena_etiquetada2)
        print("\n")
    
        file.write(cadena_etiquetada2 + os.linesep)
        
    else:
        # Printeo como queda la cadena en formato markdown

        print("Cadena formato MARKDOWN: ", cadena_etiquetada)
        print("\n")


        # Escribimos la cadena modificada en el fichero md que hemos creado

        file.write(cadena_etiquetada + os.linesep)
```

```python
# Vemos que hemos abierto bien el archivo, lineas contiene una lista con todas las cadenas
# del archivo abierto

lineas = archivo.readlines()

# Creamos el archivo markdown donde se va a escribir el codigo modificado

file = open("Martin_Carlos.md","w")

# Vamos pasando a la funcion con un bucle for cada una de las lineas del fichero HTML

for i in range(len(lineas)):
    modificaLineas(lineas[i])

# Cerramos el archivo markdown que ya hemos creado correctamente

file.close()
```
